## How much time usually passes between learning about something you need to purchase and making that purchase?

```python
bdt=ss['buy_decision_time'].value_counts()
bdt_prs=round(ss['buy_decision_time'].value_counts(normalize=True)*100, 1)
s1 = pd.Series(bdt, name='amount')
s2 = pd.Series(bdt_prs, name='percent')
buy_decision_time=pd.concat([s1, s2], axis=1)
buy_decision_time.index.name = 'buy_decision_time'
buy_decision_time.sort_values("percent", ascending=False)
```

## What’s the one thing holding you back from buying more on MROSupply?

```python
changes=ss["changes_requested"].values
print(type(changes))
changes[:10]
```

```python
new_list=[]
for c in changes:
    if "," not in c:
        new_list.append(c)
    else:
        words = c.split(",")
        for w in words:
            new_list.append(w.strip())
new_list[:20]
```

```python
def freq_table(array):
    table={}
    total=0
    for i in array:
        total+=1
        key=str(i)
        if key in table:
            table[key]+=1
        else:
            table[key]=1   
    print("total amount is: ",total)
    return table

changes_s=freq_table(new_list)
changes_s
```

```python
total=len(new_list)
changes_series = pd.Series(changes_s)
changes_requested = pd.DataFrame(changes_series, columns=['amount'])
changes_requested["percent"]=round((changes_requested["amount"]/total)*100, 1)
changes_requested.index.name = 'holding_from_buying'
changes_requested.sort_values("percent", ascending=False)
```

## If you were recommending MROsupply to a close friend or family member, what exactly would you say?

```python
import re
ss['recommendation']=ss['recommendation'].str.replace(", n/a", "")

def rep(s):
    other = re.search(r"^other", s)
    frs = re.search(r"^free shipping", s)
    price = re.search(r"^price", s)
    inv = re.search(r"^inventory", s)
    csr = re.search(r"^customer service", s)
    if other:
        s="Other"
    elif frs:
        s="Fast & Free Shipping"
    elif price:
        s="Good Price"
    elif inv:
        s="Nice Inventory"
    elif csr:
        s="Excellent Customer Service"
    return s

ss['recommendation']=ss['recommendation'].apply(rep)

recommend=ss['recommendation'].value_counts()
recommend_prs=round(ss['recommendation'].value_counts(normalize=True)*100, 1)
s1 = pd.Series(recommend, name='amount')
s2 = pd.Series(recommend_prs, name='percent')
recommendation=pd.concat([s1, s2], axis=1)
recommendation.index.name = 'recommendation'
recommendation
```
